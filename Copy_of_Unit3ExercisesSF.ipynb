## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

**Task10**:

Describe what process each of the previous discrete distributions models, and their support.

Hint: Make sure to read any of the preliZ documentation, the wikipedia articles, or wolfram alpha.


# Bernoulli:
Models probability in a binary scenario(either 0 or 1). The support is $[0,1]$

# Binomial:
Models the probability of a certain outcome occuring over a number of binary trials $n$ given a certain probability $p$. The support is $[0,n]$

# Poisson
Models probability of a certain number of events occuring over a fixed interval given the average rate of occurance $λ$. Support is $[0, \infty]$

# Geometric
Models probability that the first sucess in a series of Bernoulli trials with probability $p$ will occur on a given trial. Support is $[1,∞]$

# HyperGeometric
Models probability of successes amongst $n$ trials, given a population of $N$ Bernoulli trials, given $k$ successes, when sampling $n$ times without replacement. Support is $[max(n-(N-k),0),min(n,k)]$

# Categorical
Models probability of a certain integer given an array of probabilities $p[x]$ such that all float values of $x$ add up to 1, where the index of probabilities $x$ correspond to the probability of the integer they represent. Support is $len(p)$
